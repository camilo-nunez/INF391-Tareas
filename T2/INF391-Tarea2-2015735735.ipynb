```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from collections import Counter
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet
from scipy.spatial.distance import pdist
from sklearn.cluster import AgglomerativeClustering
```

```python
data = pd.read_csv('shopping-data.csv', delimiter=',')
```

```python
data.head()
```

Los INGRESOS están en miles de dólares y GASTOS es un número entre 1 y 100 que indica con qué frecuencia una persona gasta dinero en un centro comercial.

**Objetivos**
* Generar un dendrograma y realizar un corte para formar clusters. 
* Graficar los datos agrupados y analizar los clusters encontrados. 

**Preguntas**
* ¿Qué tipo de consumidor pertenece a cada cluster?.
* ¿Qué grupo genera mayor interés para un local de comercio?. 
* ¿Tiene algo que ver el género de una persona con el gasto?. Una visualización que podría ayudar es usar la etiqueta GENERO como marca del punto.

## Desarrollo

Para comenzar el desarrollo, definiremos el array de pares ordenados a usar, a partir de las columnas _'GASTOS'_ e _'INGRESOS'_ del dataframe.

```python
X=data[['GASTOS','INGRESOS']].to_numpy()
```

Antes de cualquier análisis debemos considerar la disponibilidad de los puntos en el plano cartesiano, y observar de forma rápida su distribución. Para ello haremos un scatter plot de los datos, etiquetados de acuerdo al género.

```python
_, (ax1) = plt.subplots(1,1, figsize=(12,12), dpi=120)
ax1.scatter(X[:,0],X[:,1], c='DarkBlue',alpha=0.5)

for i, txt in enumerate(data.GENERO):
    ax1.annotate(txt, (data.GASTOS.iat[i],data.INGRESOS.iat[i]))

ax1.set_title('shopping-data.csv')
plt.xlabel('GASTOS')
plt.ylabel('INGRESOS')

plt.tight_layout()
plt.show()
```

Para comenzar a desarrollar nuestro dendrograma, debemos contruir la matriz de adyacencia $Z$ de las distancia entre pares. Para ello, hemos de definir una método de distancia para calcular.

Dado que en la figura no tenemos alguna medida de tendencia de forma natural, buscaremos cual es la que mejor se acomoda entre las vistas en clases y en el texto guía, las cuales son 'single','complete','average','centroid' y 'ward'.

El indice que utilizaremos para definir si un método es mejor que otro, será el coeficiente de correlación cophenetic. Este indice mide la correlación entre las distancias iniciales, tomadas a partir de los datos originales, y las distancias finales obtenidas por un método seleccionado. Esta métrica es similar a la de Pearson, por lo que el mayor valor nos indicará una mejor calidad del método.

```python
for meth_i in ['single','complete','average','centroid','ward']:
    aux_Z = linkage(X, meth_i)
    aux_c, _ = cophenet(aux_Z, pdist(X))
    print(meth_i,aux_c)
```

Podemos obserevar que la mejor métrica fue alcanza por _Single Link_ , sin embargo, esta técnica es sensible a ruidos o outliers que afecten los datos, los cuales se encuentran presentes en nuestro dataset, por lo que el siguiente mejor método corresponde a _Group Average_.

```python
meth='average'
```

```python
Z = linkage(X, meth)
```

Apartir de la matriz de adyacencia $Z$, podemos calcular el dendograma y graficarlo.

```python
plt.figure(figsize=(12,12), dpi=120)
plt.title('Hierarchical Clustering Dendrogram - shopping-data.csv')
plt.ylabel('%s distance'%meth)
dendrogram(Z,
          truncate_mode='lastp',
          show_leaf_counts=False,
          leaf_rotation=90.,
          leaf_font_size=12.,
          show_contracted=True,
)
plt.axhline(y=30)
plt.show()
```

El corte que nos determinará los $K$ clusters lo definiremos con la recta $y=30$.

```python
k=7
```

A continuación, generamos el modelo que tendrá nuestros $K$ clusters, calculados por medio de método _Group Average_.

```python
model = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage=meth)
model.fit(X)
```

Finalmente, disponemos de un nuevo scatter plot pero ahora con nuestros clusters identificados.

```python
_, (ax2) = plt.subplots(1,1, figsize=(12,12), dpi=120)
scatter =  ax2.scatter(X[:,0],X[:,1], c=model.labels_,alpha=0.5,cmap='rainbow')

legend1 = ax2.legend(*scatter.legend_elements(),
                    loc="upper right", title="ID Clase")
ax2.add_artist(legend1)

for i, txt in enumerate(data.GENERO):
    ax2.annotate(txt, (data.GASTOS.iat[i],data.INGRESOS.iat[i]))

ax2.set_title('shopping-data.csv')
plt.xlabel('GASTOS')
plt.ylabel('INGRESOS')

plt.tight_layout()
plt.show()
```
