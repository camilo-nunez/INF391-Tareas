```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import genfromtxt

from sklearn import metrics
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
```

```python
# Pureza
def purity_score(y_true, y_pred):   
    contingency_matrix = metrics.cluster.contingency_matrix(y_true, y_pred)   
    return np.sum(np.amax(contingency_matrix, axis=0)) / np.sum(contingency_matrix) 
```

## Acerca de Métricas

Para el desarrollo de esta tarea, se harán usos de las métricas de evaluación `Mean Silhouette Coefficient` y `Purity Score`. Antes de aplicar cada una de estas métricas en los respectivos data set, debemos entender como se componen.

La evaluación de un cluster, para nuestros propósitos, se puede aplicar a dos tipos, `Unsupervised` y `Supervised`. 

En el caso de `Unsupervised` nos fijaremos en la `chohesion` de un cluster, la cual determina que tan cerca están los objetos en un cluster; y en la `separation`, la cual nos indicará que tan distante o bien separado está un cluster respecto de los otros. El estadístico que nos entregara una combinación de estas métricas será el `Mean Silhouette Coefficient`. Los valores de este coeficiente varían entre $[-1,1]$. Un valor negativo es indeseable, dado que nos indica el caso en que el valor de la distancia promedio de un objeto con respecto a los otros objetos del mismo cluster es **mayor** que la mínima distancia promedio del punto con respecto a otros puntos de otros clusters. Mientras el valor sea más cerca de 1, nos indica que la distancia promedio de un objeto con respecto a los otros objetos del mismo cluster tiende a 0.


En el caso de `Supervised`, vamos a considerar una técnica del tipo `classification-oriented`, el cual evalúa el comportamiento de las `single class` de los objetivos en un cluster. La técnica a utilizar será `Purity Score`, la cual nos indica el grado que tienen los objetos de estar en una `single class`.

## Data 1

```python
data_1 = genfromtxt('pathbased.txt', delimiter='\t')
clusters = len(set(data_1[:,2]))
print ('Número de clusters:', clusters)
```

```python
kmeans = KMeans(n_clusters=clusters, random_state=0).fit(data_1[:,0:2])
hac = AgglomerativeClustering(n_clusters=clusters, linkage='single').fit(data_1[:,0:2])
dbscan = DBSCAN(eps=2, min_samples=5).fit(data_1[:,0:2])

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(12,3), dpi=120)
ax1.scatter(data_1[:, 0], data_1[:, 1], s=5, c= 'black')  
ax2.scatter(data_1[:, 0], data_1[:, 1], s=5, c= kmeans.labels_)  
ax3.scatter(data_1[:, 0], data_1[:, 1], s=5, c= hac.labels_)  
ax4.scatter(data_1[:, 0], data_1[:, 1], s=5, c= dbscan.labels_)  
ax1.set_title('Original')
ax2.set_title('K-Means')
ax3.set_title('Single-link')
ax4.set_title('DBSCAN')
plt.tight_layout()
plt.show()
```

### Comentarios

A simple vista, de las tres técnicas usadas, el método de `K-Means` muestra una distribución más exacta de los datos, incluso, considerando su desventaja de tender a hacer globular-shaped clusters en los datos. A diferencia de los otros dos métodos, que no muestran una clusterización ideal.

### Métricas

```python
silhouette_score_arr = []
silhouette_score_arr.append(metrics.silhouette_score(data_1, kmeans.labels_, metric='euclidean'))
silhouette_score_arr.append(metrics.silhouette_score(data_1, hac.labels_, metric='euclidean'))
silhouette_score_arr.append(metrics.silhouette_score(data_1, dbscan.labels_, metric='euclidean'))

purity_score_arr = []
purity_score_arr.append(purity_score(data_1[:, 2], kmeans.labels_))
purity_score_arr.append(purity_score(data_1[:, 2], hac.labels_))
purity_score_arr.append(purity_score(data_1[:, 2], dbscan.labels_))

y = ['K-Means', 'HAC/Single-link', 'DBSCAN']
y_pos = [i for i, _ in enumerate(y)]

plt.rcdefaults()
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6), dpi=120)

ax1.barh(y_pos, silhouette_score_arr, align='center')
ax1.set_yticks(y_pos)
ax1.set_yticklabels(y)
ax1.set_xticks([-0.2, -0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
ax1.invert_yaxis()
ax1.set_xlabel('Silhouette Coeff [-1,1]')
ax1.set_title("Mean Silhouette Coefficient")
for i, v in enumerate(silhouette_score_arr):
    ax1.text(v, i, " "+str(v), va='baseline') 

ax2.barh(y_pos, purity_score_arr, align='center')
ax2.set_yticks(y_pos)
ax2.set_yticklabels(y)
ax2.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
ax2.invert_yaxis()
ax2.set_xlabel('Putiry [-1,1]')
ax2.set_title("Purity")
for i, v in enumerate(purity_score_arr):
    ax2.text(v, i, " "+str(v), va='baseline') 

plt.show()
```

### Análisis

De acuerdo al uso de las métricas presentadas anteriormente, y de acuerdo a los datos graficados en la figura anterior, podemos observar que la técnica de `K-Means` se confirma que tuvo un mejor desempeño de entre las otras dos. Dado que de los tres coeficiente `Silhouette`, el de `K-Means` es el más cercano a 1, lo cual nos indica que la distancia promedio de un objeto con respecto a los otros objetos en el cluster de `K-Means` tienden a 0.

Quien tuvo el peor desempeño fue `Single-link`. Lo cual tiene coherencia si se observa la gráfica del cluster dado.

## Data 2

```python
data_2 = genfromtxt('Compound.txt', delimiter='\t')
clusters = len(set(data_2[:,2]))
print ('Número de clusters:', clusters)
```

### Comentarios

A simple vista, el método con `DBSCAN` posee una mejor representación de los datos. Luego le sigue `K-Means`, quien si bien tiene un buen aspecto, genero globular-shaped en sus representaciones, segmentando los datos de las figuras. Finalmente, `Single-link` se espera que obtenga el peor resultado.

### Métricas

```python
silhouette_score_arr = []
silhouette_score_arr.append(metrics.silhouette_score(data_2, kmeans.labels_, metric='euclidean'))
silhouette_score_arr.append(metrics.silhouette_score(data_2, hac.labels_, metric='euclidean'))
silhouette_score_arr.append(metrics.silhouette_score(data_2, dbscan.labels_, metric='euclidean'))

purity_score_arr = []
purity_score_arr.append(purity_score(data_2[:, 2], kmeans.labels_))
purity_score_arr.append(purity_score(data_2[:, 2], hac.labels_))
purity_score_arr.append(purity_score(data_2[:, 2], dbscan.labels_))

y = ['K-Means', 'HAC/Single-link', 'DBSCAN']
y_pos = [i for i, _ in enumerate(y)]

plt.rcdefaults()
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,6), dpi=120)

ax1.barh(y_pos, silhouette_score_arr, align='center')
ax1.set_yticks(y_pos)
ax1.set_yticklabels(y)
ax1.set_xticks([-0.2, -0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
ax1.invert_yaxis()
ax1.set_xlabel('Silhouette Coeff [-1,1]')
ax1.set_title("Mean Silhouette Coefficient")
for i, v in enumerate(silhouette_score_arr):
    ax1.text(v, i, " "+str(v), va='baseline') 

ax2.barh(y_pos, purity_score_arr, align='center')
ax2.set_yticks(y_pos)
ax2.set_yticklabels(y)
ax2.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
ax2.invert_yaxis()
ax2.set_xlabel('Putiry [-1,1]')
ax2.set_title("Purity")
for i, v in enumerate(purity_score_arr):
    ax2.text(v, i, " "+str(v), va='baseline') 

plt.show()
```

### Análisis

Al observar los datos obtenidos de las métricas graficadas, se puede comprobar que el desempeño de `DBSCAN` fue el mejor de los tres, dado que posee el mejor `Mean Silhouette Coefficient`, y su `Purity Score` esta más cerca de 1, validando nuestra hipótesis. Sin embargo, hay que tomar en cuenta que las formas no globulares que tienen los datos; dado que esto es un impedimento claro para el desempeño de `K-Mean`, de acuerdo a lo mencionado en el análisis anterior. Por lo tanto, dado las formas de los datos, era de esperar que `DBSCAN` fuese el mejor método.

Con respecto a `Single-link`, no se esperaba mucho, dado que nuevamente tuvo el peor desempeño comprobado.
