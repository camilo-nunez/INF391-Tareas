1.   Ejecute **Spectral Clustering** en cada uno de los siguientes conjuntos de datos y comente en cuál de ellos es capaz de encontrar la agrupación correcta ("natural"). 

2.   Detalle una situación en la cuál sería conveniente utilizar K-Means en lugar de Spectral Clustering.

```python
from numpy import genfromtxt
from sklearn.cluster import SpectralClustering
import matplotlib.pyplot as plt
```

```python
d1 = genfromtxt('moons.csv', delimiter=',')
d2 = genfromtxt('rings.csv', delimiter=',')
d3 = genfromtxt('spiral.csv', delimiter=',')
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,3), dpi=120)
ax1.scatter(d1[:,0], d1[:,1], s=1)  
ax2.scatter(d2[:,0], d2[:,1], s=1) 
ax3.scatter(d3[:,0], d3[:,1], s=2) 
plt.tight_layout()
plt.show()
```

# Desarrollo

En esta oportunidad, para construir la matriz de afinidad se utilizará el kernel `radial basis function (RBF)`. Por otro lado, se definieron los `n_clusters` de acuerdo a los conjuntos naturales esperados.

```python
d1_model = SpectralClustering(n_clusters=2,random_state=1,affinity='rbf',gamma=50).fit(d1)
d2_model = SpectralClustering(n_clusters=3,random_state=1,affinity='rbf',gamma=100).fit(d2)
d3_model = SpectralClustering(n_clusters=3,random_state=1,affinity='rbf',gamma=10).fit(d3)
```

```python
fig, axs = plt.subplots(1,3, figsize=(12,3), dpi=120)

axs[0].scatter(d1[:, 0], d1[:, 1], s=5, c=d1_model.labels_)
axs[0].set_title("(gamma=50)\n(n_clusters=2)")
axs[1].scatter(d2[:, 0], d2[:, 1], s=5, c=d2_model.labels_)
axs[1].set_title("(gamma=100)\n(n_clusters=3)")
axs[2].scatter(d3[:, 0], d3[:, 1], s=5, c=d3_model.labels_)
axs[2].set_title("(gamma=10)\n(n_clusters=3)")

plt.tight_layout()
plt.show()
```

En los tres casos se logró encontrar la agrupación correcta. Esto se debe a que el kernel usado para calcular la matriz de afinidad fue RBF, el cual pondera  de forma gaussiana la distancia entre vectores. Luego, los puntos que están presentes en grupos con distancias más pequeñas fueron incluidos en el mismo cluster, _"forzando"_ a obtener una forma a la natural.

Con respecto a `K-means`, debemos considerar que una de sus limitaciones es que tiende a hacer globular-shaped clusters en los datos, por lo que datos con formas aleatorias y con centroides solapados no serían buenos candidatos para clusterizarlos con `K-means`, además de aquellos datos con dimensiones muy grandes. Sin embargo, cuando la dimensionalidad de los espacios vectoriales de los datos es pasada por transformaciones lineales que la van reduciendo a localidades más pequeñas o a espacios euclidianos, `K-means` se vuelve una opción mucho más viable y optima.

Luego, para datos en espacios euclidianos simples, `K-means` es mejor opcción que `Spectral Clustering`.
