```python
# Librerías útiles
import matplotlib.pyplot as plt
from numpy import genfromtxt
from sklearn.cluster import KMeans

from pandas import read_csv
```

```python
# Cargar los datos
d1 = genfromtxt('../input/T1-1.csv', delimiter=',')
d2 = genfromtxt('../input/T1-2.csv', delimiter=',')
d3 = genfromtxt('../input/T1-3.csv', delimiter=',')
d4 = genfromtxt('../input/T1-4.csv', delimiter=',')
```

```python
# Crear Figuras
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(12,3), dpi=120)

# Plot
ax1.scatter(d1[:,0], d1[:,1], s=1, marker=".")  
ax2.scatter(d2[:,0], d2[:,1], s=1) 
ax3.scatter(d3[:,0], d3[:,1], s=2) 
ax4.scatter(d4[:,0], d4[:,1], s=2)

ax1.set_title('(K=3)')
ax2.set_title('(K=3)')
ax3.set_title('(K=2)')
ax4.set_title('(K=3)')

#ax1.set_xlim(-1,6)
#ax1.set_ylim(-1,6)

#ax2.set_xlim(-1,3)
#ax2.set_ylim(-1,3)

plt.tight_layout()
plt.show()
```

```python
# Objetos para las figuras
fig_means, (c_ax1, c_ax2, c_ax3, c_ax4) = plt.subplots(1,4, figsize=(12,3), dpi=120)

# Coleccion por dataset
C_th=[(3,d1,c_ax1),(3,d2,c_ax2),(2,d3,c_ax3),(3,d4,c_ax4)]

# Ejecucion por cada coleccion
for c_i in C_th:
    kmeans_i = KMeans(n_clusters=c_i[0], init='k-means++', max_iter=300, n_init=10, random_state=0)
    
    kmeans_i.fit(c_i[1])
    pred_y_i = kmeans_i.predict(c_i[1])
    centers_i = kmeans_i.cluster_centers_

    c_i[2].scatter(c_i[1][:,0], c_i[1][:,1],c=pred_y_i, cmap='viridis')
    c_i[2].scatter(centers_i[:, 0], centers_i[:, 1], s=300, c='pink', alpha=0.5)
    c_i[2].set_title('(K=%s)' % str(c_i[0]))
```
